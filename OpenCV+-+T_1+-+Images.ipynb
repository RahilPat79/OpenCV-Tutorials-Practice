# Video  

```python
cap=cv2.VideoCapture(r'C:\Users\rahil\Desktop\Tensor Flow\videoplayback.mp4') #0 to play from camera
while(True):
    #Capture frame by frame
    ret,frame = cap.read()
    
    #Our operations on the frame
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    
    #Display the resulting frame
    cv2.imshow('frame',gray)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
#When everything is done release the capture
cap.release()
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2
cap=cv2.VideoCapture(r'C:\Users\rahil\Desktop\Tensor Flow\videoplayback.mp4')

#Define the codec and create VideoWriter
fourcc = cv2.VideoWriter_fourcc(*'DIVX')
out=cv2.VideoWriter(r'C:\Users\rahil\Desktop\Tensor Flow\new_playback.mp4',fourcc,20.0,(640,480))

while(cap.isOpened()):
    ret,frame = cap.read()
    if ret==True:
        frame=cv2.flip(frame,0)
        gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        #write the flipped frame
        out.write(gray)
        
        cv2.imshow('frame',gray)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
#Release the frame
cap.release()
out.release()
cv2.destroyAllWindows()
```

# Drawing shapes

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

img = np.zeros((512,512,3), np.uint8)

#Draw a blue Diagonal line
img=cv2.line(img,(0,0),(511,511),(255,0,0),5)
#cv2.imshow('draw',img)
#cv2.destroyAllWindows()
img= cv2.rectangle(img,(384,0),(510,128),(0,255,0),3)
img=cv2.circle(img,(447,63), 63, (0,0,255), -1)
img= cv2.ellipse(img,(256,256),(100,50),0,0,180,255,-1)
```

```python
pts = np.array([[10,5],[20,30],[70,20],[50,10]], np.int32)
pts=pts.reshape((-1,1,2))
img=cv2.polylines(img,[pts],True,(0,255,255)) # if True is False we get polylines joining all points not a closed shape
font= cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img,'Rahil Patel',(10,500),font,4,(255,255,255),2),cv2.LINE_AA
cv2.imshow('image',img)
plt.imshow(img)
plt.xticks([]) , plt.yticks([])
plt.show()
```
